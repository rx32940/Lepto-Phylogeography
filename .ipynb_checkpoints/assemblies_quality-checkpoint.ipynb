```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
seq_path = "/Users/rx32940/Desktop"

species_table = pd.read_csv(seq_path + "/species_list/Leptospira interrogans", sep=",")
species_acc = species_table['V1']
```

```python

biosamples = pd.DataFrame()

for file in species_acc:
    report = pd.read_csv(seq_path+"/All_QUAST/" + file + ".tsv", sep= "\t")
    report['biosample_acc'] = file
    biosamples = biosamples.append(report)
    
    
```

```python
biosamples
```

```python
N50 = biosamples['N50'].describe()
num_contigs = biosamples['# contigs'].describe()
num_misasm = biosamples['# misassemblies'].describe()

good_N50 = biosamples[biosamples.N50 > biosamples.N50.quantile(.95)]
good_length = biosamples[biosamples['Total length'] > biosamples['Total length'].quantile(.75)]
less_misasm = biosamples[biosamples['# misassemblies'] < biosamples['# misassemblies'].quantile(.05)]
intersect = set(good_N50['biosample_acc']).intersection(set(less_misasm['biosample_acc']))


print(intersect)
# N50.plot.kde()
```
