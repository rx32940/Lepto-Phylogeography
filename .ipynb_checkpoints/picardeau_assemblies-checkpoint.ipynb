```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET
import re
```

```python
path="/Users/rx32940/Dropbox/5.Rachel-projects/Phylogeography/Lepto_assemblies_V2/picardeau/"

doc = ET.parse(path + "picardeau_313_biosamples_metadate.xml") # xml object, biosample result file downloaded directly from ncbi with picardeau_313_biosample list, format with vscode tool

root = doc.getroot() # get the root element

all_picardeau = pd.DataFrame()


for biosample in root.findall("BioSample"):
    biosample_acc = biosample.attrib["accession"]
    sample_name = biosample.find("Description").find("Title").text
    biosample_name = biosample.find("Description").find("Organism").find("OrganismName").text
    species=' '.join(biosample_name.split(" ")[:2])
    metadata = pd.DataFrame({"Biosample_acc":[biosample_acc], "sample_name":[sample_name], "biosample_name": [biosample_name],"species":[species]})
    all_picardeau = all_picardeau.append(metadata)
    
species_dict=all_picardeau[["Biosample_acc","species"]]
species_dict.to_csv(path+"biosample_species_dict_picardeau.csv",header=None,index=False)
species_dict
#print(root)
```
