### NCBI Assemblies Summary
   
   - N50
   - Total Length
   - num of contigs

```python
# N50

ax= sns.distplot(ncbi_isolates["N50"],hist=True,kde=False, bins=int(2000000/10000),color="r")
ax2 = ax.twinx()
sns.distplot(ncbi_isolates['N50'],hist=False,kde=True,bins=int(2000000/10000),color="g")
plt.title("NCBI Downloaded Assemblies N50 Distribution",fontsize=15,y=1.08)
ax.set_ylabel("Histogram")
ax2.set_ylabel("KDE")

for tick in ax.get_xticklabels():
    tick.set_rotation(45)
```

```python
ax= sns.distplot(ncbi_isolates["Total length"],hist=True,kde=False, bins=int(5500000/10000),color="r")
ax2 = ax.twinx() # subplot of ax
sns.distplot(ncbi_isolates['Total length'],hist=False,kde=True,bins=int(5500000/10000),color="g")
plt.title("NCBI Downloaded Assemblies Scaffold Length Distribution",fontsize=15,y=1.08)
ax.set_ylabel("Histogram")
ax2.set_ylabel("KDE")
labels = ax,ax2.get_xticklabels()

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

```

```python
ax= sns.distplot(ncbi_isolates["# contigs"],hist=True,kde=False, bins=int(600),color="r")
ax2 = ax.twinx() # subplot of ax
sns.distplot(ncbi_isolates['# contigs'],hist=False,kde=True,bins=int(600),color="g")
plt.title("NCBI Downloaded Assemblies Number of Contigs Distribution",fontsize=15,y=1.08)
ax.set_ylabel("Histogram")
ax2.set_ylabel("KDE")
labels = ax,ax2.get_xticklabels()

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

```

```python
ax = sns.distplot(self_asm_isolates["N50"],hist=True,kde=False,bins=int(2000000/10000), color="r")
ax2 = ax.twinx()
sns.distplot(self_asm_isolates["N50"],hist=False,kde=True,bins=int(2000000/10000),ax=ax2,color="g")
plt.title("Self_assembled Biosamples N50 Distribution",fontsize=15,y=1.08)
ax.set_ylabel("Histogram")
ax2.set_ylabel("KDE")
```

```python
ax = sns.distplot(self_asm_isolates["Total length"].astype(int),hist=True,kde=False,bins=int(5500000/10000), color="r")
ax2 = ax.twinx()
sns.distplot(self_asm_isolates["Total length"].astype(int),hist=False,kde=True,bins=int(5500000/10000),ax=ax2,color="g")
plt.title("Self_assembled Biosamples Scaffold Length Distribution",fontsize=15,y=1.08)
ax.set_ylabel("Histogram")
ax2.set_ylabel("KDE")
ax.ticklabel_format(style='plain')

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

```

```python
ax = sns.distplot(self_asm_isolates["# contigs"].astype(int),hist=True,kde=False,bins=int(600), color="r")
ax2 = ax.twinx()
sns.distplot(self_asm_isolates["# contigs"].astype(int),hist=False,kde=True,bins=int(600),ax=ax2,color="g")
plt.title("Self_assembled Biosamples Number of Contigs Distribution",fontsize=15,y=1.08)
ax.set_ylabel("Histogram")
ax2.set_ylabel("KDE")
ax.ticklabel_format(style='plain')

for tick in ax.get_xticklabels():
    tick.set_rotation(45)
```
