### Coverage of Self Assembled Isolates

- **coverage** of the sequences were calculated by mapping short reads to corresponding reference genome with **BWA** and calculated with **Qualimap**

- get coverage info from qualimap analysis (mind that this still **includes the isolates with low genome fraction** filtered out from last section)

```python
# picardeau multiqc qualimap report
picardeau_qualimap = pd.read_csv(path + "/picardeau/multiqc_qualimap_general_stats_picardeau.txt",sep="\t")
picardeau_cov_filtered = picardeau_qualimap#[(picardeau_qualimap["QualiMap_mqc-generalstats-median_coverage"]>= 20) # median coverage at least 30X
                                            #& (picardeau_qualimap["QualiMap_mqc-generalstats-30_x_pc"] >=50)][["Sample","QualiMap_mqc-generalstats-median_coverage"]]# at least 50% of the scaffold is >= 30X
picardeau_cov_filtered

# dated multiqc qualimap report
date_qualimap = pd.read_csv(path + "/date/multiqc_qualimap_general_stats_date.txt",sep="\t")
date_cov_filtered = date_qualimap#[(date_qualimap["QualiMap_mqc-generalstats-median_coverage"]>= 20) # median coverage at least 30X
                                 #           & (date_qualimap["QualiMap_mqc-generalstats-30_x_pc"] >=50)][["Sample","QualiMap_mqc-generalstats-median_coverage"]]# at least 50% of the scaffold is >= 30X
date_cov_filtered

rest_qualimap = pd.read_csv(path + "/rest/multiqc_qualimap_general_stats_rest.txt",sep="\t")
rest_cov_filtered = rest_qualimap#[(rest_qualimap["QualiMap_mqc-generalstats-median_coverage"]>= 20) # median coverage at least 30X
                                 #           & (rest_qualimap["QualiMap_mqc-generalstats-30_x_pc"] >=50)][["Sample","QualiMap_mqc-generalstats-median_coverage"]]# at least 50% of the scaffold is >= 30X
rest_cov_filtered

All_sample_coverage = pd.concat([picardeau_cov_filtered,date_cov_filtered,rest_cov_filtered], ignore_index=True, sort=False)[["Sample","QualiMap_mqc-generalstats-median_coverage"]].rename(columns={"QualiMap_mqc-generalstats-median_coverage":"coverage"})

All_sample_coverage
```

```python
# add an extra one did get mapped properly

addition_cov = pd.DataFrame({"Sample": ["SAMEA864115"], "coverage":[14.2493]})
All_sample_coverage = pd.concat([All_sample_coverage, addition_cov], ignore_index=True, sort=False)

All_sample_coverage
```

### Merge coverage with Quast Analysis report

```python
all_data_w_cov = pd.merge(All_sample_coverage,data_subset,how="right")

all_data_w_cov

all_data_w_cov.to_csv(path + "/all_data_w_cov.csv")
```

### Re-quast Scaffolds with Low Genome Fraction

- evaulate scaffolds with **genome fraction lower than 70** from QUAST report
- low genome fraction isolates were checked on NCBI with the tool **STAT** for **mis-identified species**
  - keep those with STAT classified **at least 70% reads** to the genus _Leptospira_
- requast these isolates with **STAT identified species's reference genome** (if identified as _Leptospira sp._, use _L. interrogans_)
- concate those with high genome fraction identified with the original dataframe

```python
# filtered out isolates with high mapping genome fraction
genome_frac = all_data_w_cov[all_data_w_cov["Genome fraction (%)" ] > 70] 
```

```python
# low mapping isolates was identified, searched on NCBI with STAT identified new species, re-quasted
low_map_picardeau = pd.read_csv(path+"/picardeau/quast/low_map_multiqc_quast.txt", sep="\t")
low_map_rest = pd.DataFrame(pd.read_csv(path + "/rest/low_map_multiqc_quast.txt",sep="\t"))

# concate re-quasted isolates from two datasets, excluding SAMN01919665
low_map = pd.concat([low_map_picardeau,low_map_rest],sort=False, ignore_index=True).loc[0:8,:]
low_map["Sample"]=low_map["Sample"].str.split(" | ").str[0]

# reformat the low-map data frame so it can concatenated with the orginal data frame
low_map_subset = low_map[["Sample","coverage","N50","# genomic features","# contigs","Genome fraction (%)","Total length"]]
```

```python
# concate orginal high genome frac isolates, re-mapped&quasted islates together
all_self_assemblies = pd.concat([genome_frac,low_map_subset], ignore_index=True, sort=False)
all_self_assemblies_filtered= all_self_assemblies[(all_self_assemblies["coverage"] >= 20) &
                                                 (all_self_assemblies["# contigs"] <= 500) &
                                                 (all_self_assemblies["N50"] >= 10000)]
# in addition to these filter above, genome fraction from quast must be higher than 70%
all_self_assemblies_filtered
```

```python
all_self_assemblies_filtered.to_csv(path + "/all_self_assemblies_filtered.csv",header=True, index=False)
```
